```python
import pandas as pd
data = pd.read_csv('report.txt', sep = " ")#seperated by space!
data.head()
```

# Calculate sum and average across columns
- df.columns: get the name of each
- df.columns[index]: get specific name by index
- df.describe(): get the basic info of each column (mean...)
- df.iloc[:,1:]: select specific columns (subset) by index
- df.sum(axis=1): sum over columns. axis=0 means row; axis=1 means column
- df.mean(axis=1)
- df["new_column_name"]=value: append new column
- df_new = df.round(decimals = 3): keep the decimals place as 3. the change needs to be saved to new variable.

```python
data1 = data.iloc[:,1:] 
# calculate the total and mean score for each subject
col_sum = data1.sum(axis = 1) #sum over columns
col_avg = data1.mean(axis = 1)
data["总分"] = col_sum
data["平均分"] = col_avg
data.head()
```

# Calculate sum and average across rows
- df.loc[index] = value: append one row value in the specific place.
- how to insert one row on the top:
 - df.loc[-1] = value: append the row value at very last
 - df.index = df.index + 1: make all index + 1, so the index of last row is 0
 - df.sort_index(inplace = True):sort by index, make the last row to the top since the row organization order is 0,1,2,3...
- df.loc[repace_specific_row_by_boolean,"column_name"] = value: value specific cell

```python
data2 = data.iloc[:,1:] 
row_avg = data2.mean(axis = 0) #mean over rows
# # insert a row at very first
data.loc[-1] = row_avg #insert row at last
data.index = data.index + 1 #organize the index
data.sort_index(inplace=True) #put index=0 to very first
data.loc[0, '姓名'] = "平均" #modify specific cell value
data.head()
```

# Replace value under condition

```python
for i in range(1,10):
    data.loc[data.iloc[:,i] < 60, data.columns[i]] = "不及格"
data.head()
```

# Rank by rules
- df[specific_column_name].where(rank_condition).rank(). rank by specific column (df[specific_column_name])+rank under some conditions(where(rank_condition))+rank method
- df[[name_you_want_put_on_first_column]+df.columns[:-1].tolist()]: organize the order of column
- df.sort_values(by=[sort_by_which_column], inplace=True): sort by value

```python
data['名次'] = (data['总分'].where(data.姓名 != "平均").rank(method='max',ascending = False))
data = data[['名次'] + data.columns[:-1].tolist()]
data.sort_values(by=['名次'],inplace = True)
data.head()
```

# Organize and output
- insert the row without using index (because we have already make the sort so we don't want to re-do)
 - data_shifted = data.shift(1): add new row on the first with NaN
 - data_shifted.iloc[0,:] = data.iloc[-1,:]: value the first NaN row
- df.round({"column_name1":1, "col_name2":2}): round with specific decimal places

```python
data_shifted = data.shift(1)
data_shifted.iloc[0,:] = data.iloc[-1,:]
data_shifted.loc[14, '名次'] = 0 #pay attention to the difference on index between df.iloc() and df.loc
data = data_shifted

data = data.round({"平均分":1}) 

data.to_csv(r'result.txt')
```
